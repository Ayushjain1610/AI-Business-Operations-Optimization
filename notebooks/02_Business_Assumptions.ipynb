```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/ayushjain/Documents/AI-Powered Business Operations & Profit Optimization System/data/clean_data.csv")
df.head()
```

```python
df.rename(columns={
    'transaction_date': 'Date',
    'product_name': 'Product',
    'quantity': 'Units_Sold'
}, inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Price'] = df['final_amount'] / df['Units_Sold']
```

```python
df['Unit_Cost'] = df['Price'] * 0.45
df['Holding_Cost'] = df['Unit_Cost'] * 0.02
df['Shortage_Cost'] = df['Unit_Cost'] * 0.10
```

```python
df['Lead_Time'] = np.random.randint(2,7,size=len(df))
```

```python
df['Capacity'] = df.groupby('Product')['Units_Sold'].transform('max') * 1.2
```

```python
df['Revenue'] = df['Units_Sold'] * df['Price']
```
