```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
forecast = pd.read_csv("../data/forecast.csv")
inventory = pd.read_csv("../data/inventory_plan.csv")
production = pd.read_csv("../data/production_plan.csv")
profit = pd.read_csv("../data/profit_results.csv")

production.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(forecast['Date'], forecast['Forecasted_Demand'])
plt.title("Demand Forecast")
plt.xlabel("Date")
plt.ylabel("Forecasted Units")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=inventory, x='Product', y='Safety_Stock')
plt.title("Safety Stock by Product")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=production, x='Product', y='Optimal_Production')
plt.title("Optimal Production Quantity")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=profit, x='Product', y='Profit')
plt.title("Profit Contribution by Product")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(profit['Product'], profit['Revenue'], label='Revenue')
plt.bar(profit['Product'], profit['Production_Cost'], label='Production Cost')
plt.legend()
plt.xticks(rotation=45)
plt.title("Revenue vs Production Cost")
plt.show()
```
